The given data represents information about opportunities in the Insurance Market for different clients, as well as information about the producers and the expected decision date. The data includes 10 Attributes:
*	Date: The date when the opportunity was first encountered by the primary producer.


*	Primary Producer: The person responsible for managing the opportunity. 


*	Stage Name: The stage of the sales process that the opportunity is in. 


*	Account Name: The name of the potential client. 


*	Opportunity Name: The name of the specific opportunity. 


*	Niche Affiliations: The industry or market segment that the potential client belongs to. 


*	Expected Decision Date: The expected date when the potential client will make a decision about whether or not to work with the insurance company. 


*	Annual Revenue: The estimated annual revenue of the potential client. 


*	Acct ID: The account ID of the potential client. 


*	Primary Office: The office responsible for managing the opportunity. 

    
> We will further analyse the performance of the primary producers, identify potential patterns or trends in the sales process, and make strategic decisions based on the characteristics of the potential clients.


## Importing Libraries

```python
import pandas as pd                             #importing pandas as pd
import numpy as np                              #importing numpy as np
import seaborn as sns                           #importing seaborn as sns
import matplotlib.pyplot as plt                 #importing matplotlib as plt
```

## To read excel Dataset
We have to import our data into the jupyter noterbook in order to analyze. We will use pandas to read xlsx file and to store it in a dataframe as follws:

```python
## reading Dataset
df =pd.read_excel('Insurancedata.xlsx')
df.head()
```

## Extracting Tables

From above we can see there are three various tables are present in the dataset we have to separate the table in order to consolidate the data.

```python
#extracting table 1 and save it in df1
df1 = df.iloc[:,0:8]               
df1
```

```python
#extracting table 2 and save it in df2
df2 = df.iloc[0:5,9:11]

#renaming the column names
df2 = df2.rename(columns={'Primary Producer.1': 'Primary Producer', 'Office' : 'Primary Office'})
df2.head()
```

```python
#extracting table 3 and save it in df3
df3 = df.iloc[0:77,14:]

#renaming the column names
df3 =df3.rename(columns = {'Account Name.1':'Account Name'})
df3
```

## Consolidating Dataset

We will merge the Table1 and Table3 in order to consolidate our data together and to find insights. We will merge dataframes as follows:

```python
df_final = pd.merge(df1, df3, on ='Account Name', how='left')  #merging dataframes
df_final
```

## Basic exploration of the dataset

```python
#To check for basic information of the dataset such as null elements and datatypes of column, memory usage by data
df_final.info()
```

From above we can observe that all the dtypes are taken as object except Annual Revenue. We will further **change the data type of the 'Date' and 'Expected Decision Date' to datetime for our analysis purpose**.

```python
df_final['Date'] = pd.to_datetime(df_final['Date'], format='%d/%m/%Y', errors='coerce')
df_final['Expected Decision Date'] = pd.to_datetime(df_final['Expected Decision Date'], format='%d/%m/%Y', errors='coerce')
```

## Dealing with Missing and Duplicate values

Handling missing values and duplicate values is the important step to perform before analysis of data. Presence of this kind of data can affect our analysis and create bias in the study of dataset.

```python
#to check which column have missing values using isna()
df_final.isna().sum()
```

We can observe there are **null values present in 3 in Date, 18 in Niche Affiliations, 2 in Expected Decision Date and 6 in Annual Revenue. We will explore the null values further**.

```python
df_final[df_final.isna().any(axis=1)]       #exploring null values
```

>We can drop the null values present in the Date, Expected Decision Date and Annual Revenue as it is very less in number and it can create bias in our analysis. But we will not drop the null values present in Niche Affiliations. We will replace it with Unknown value.

```python
# Drop all rows with NaNs in Annual Revenue
df_final.dropna(subset=['Annual Revenue'], inplace = True)
#to check which column have missing values using isna()
df_final.isna().sum()
```

```python
df_final['Niche Affiliations'].fillna('Unknown', inplace =True)         #replacing null with Unknown value.
```

```python
df_final.dropna(inplace=True)       #to drop all null values
```

```python
df_final.isna().sum()               #to check if any null values present ot not
```

We have  successfully handled the null values present in the data set. now we will check for **duplicates**.

```python
df_final.duplicated().sum()
```

```python
#dropping all duplicates from df
df_final.drop_duplicates(inplace=True)
df_final                                     #our final dataset
```

```python
#To check for basic information of the dataset such as null elements and datatypes of column, memory usage by data
df_final.info()
```

## Exploratory Data Aanalysis Using Visuals

We will further analyse the performance of the primary producers, identify potential patterns or trends in the sales process, and make strategic decisions based on the characteristics of the potential clients using various visalization. We will use Matplotlib and Seaborn.

### Various Stages and Number of Opportunities 

We will find out the stages of of the sales process and number of opportunities using barplot as follows:

```python
#Check the number of opportunities by stage name 
stage_counts = df_final['Stage Name'].value_counts()

sns.set(style='darkgrid')

plt.figure(figsize=(11,5))   #set fig size

#to plot barplot using seaborn
sns.barplot(x=stage_counts.index, y=stage_counts.values, palette='Paired')
plt.title('Number of Opportunities by Stage Name')
plt.xlabel('Stage Name')
plt.ylabel('Count')
plt.show()
```

The stage name column in the dataset represents the stage of the sales process that an opportunity has reached. The dataset has three distinct values for stage name:

**1-Met Client, Data Gather, Insurer Market**: This is the early stage of the sales process where the primary producer has met with the potential client and is gathering data to understand the client's needs and identify the appropriate insurance market for them.

**2-Client Presentation-Await Feedback**: This stage represents the middle stage of the sales process where the primary producer has presented a proposal to the client and is awaiting feedback from them.

**3-Closed Won**: This stage represents the final stage of the sales process where the opportunity has been successfully closed, and the insurer has won the client.

From above we can conclude:

* **Most of the opportunity are Closed Won** which means opportunity has been successfully closed, and the insurer has won the client i.e around 37

* Around **26 opportunities are in initail state** i.e. 1-Met Client, Data Gather, Insurer Market

* Around **23 opportunities are in middle state** i.e. 2-Client Presentation-Await Feedback

 ### Top 10 Accounts Having Highest Annual Revenue

### Bottom 10 Accounts Having Lowest Annual Revenue

### Count of Expected Decisions per Month (2019)

Using this information we will find out count of Expected Decisions of client per Month. This will be helpful to take the Business desicions by understanding in which month is crutial and provide clarity about further stage.

### Annual Revenue vs. Expected Decisions per Month (2019)

We will try to find out for each month Total Annual Revenue and Expected Decisions per Month (2019).


```python
# Group the data by Primary Producer and Expected Decision Month and calculate the sum of Annual Revenue
grouped_exp = df_final.groupby([df_final['Primary Producer'], df_final['Expected Decision Date'].dt.month])['Annual Revenue'].sum()

# Convert the series to a dataframe and reset the index
grouped_exp = grouped_exp.reset_index()

# Create the bar plot using seaborn
sns.barplot(data=grouped_exp, x='Expected Decision Date', y='Annual Revenue')

# Format the plot
plt.title("Annual Revenue vs. Expected Decisions per Month (2019)")
plt.xlabel("Expected Decision in Month")
plt.ylabel('Annual Revenue')
plt.show()
```

* We can observe **12th month is important in terms of Total Annual Income and Expected Decisions.**

### Total Revenue by Quarter and Expected Decisions

We will find out Quarterly Total Annual Revenue and Expected Decisions per Quarter .

```python
#create a new column 'Quarter' by extracting the quarter from the 'Expected Decision Date' column
df_final['Quarter'] = df_final['Expected Decision Date'].dt.quarter

#Group the data by quarter and find the total revenue for each quarter
quarterly_revenue = df_final.groupby('Quarter')['Annual Revenue'].sum()

#Create a bar plot
sns.barplot(x=quarterly_revenue.index, y=quarterly_revenue.values)
plt.title('Total Revenue by Quarter')
plt.xlabel('Quarters')
plt.ylabel('Total Revenue')
plt.show()
```

* **2nd Quater is having high total revenue i.e. arround 220000 GBP**

### Trend of Annual Revenue by Primary Producer Per Year

WE will find out the trend of annual revenue over time for each Primary Producer.

```python
# Group by Primary Producer and year and compute the sum of Annual Revenue
grouped = df.groupby([df_final['Primary Producer'], df_final['Date'].dt.year])['Annual Revenue'].sum().reset_index()

# Pivot the data to create a wide format for plotting
wide_df = grouped.pivot(index='Date', columns='Primary Producer', values='Annual Revenue')

# Plot a line chart for each Primary Producer
wide_df.plot(kind='line', figsize=(10, 5))
plt.xlabel('Year')
plt.ylabel('Annual Revenue')
plt.title('Annual Revenue by Primary Producer Per Year')
plt.show()
```

We can see that **all producers are trending upward in terms of Annual Revenue over period**.
